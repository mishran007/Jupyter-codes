```python
# Import data file.
import pandas as pd
from ipywidgets import *
from tkinter import Tk, filedialog
from IPython.display import clear_output, display
```

```python
def select_files(b):
    clear_output()                                         # Button is deleted after it is clicked.
    root = Tk()
    root.withdraw()                                        # Hide the main window.
    root.call('wm', 'attributes', '.', '-topmost', True)   # Raise the root to the top of all windows.
    b.files = filedialog.askopenfilename(multiple=False)    # List of selected files will be set button's file attribute.
    print(b.files)                                         # Print the list of files selected.
    
fileselect = Button(description="Select File")
fileselect.on_click(select_files)

display(fileselect) # Read one of the 'div' output files.
```

```python
# Call the file in the above location 'file'. 
file = fileselect.files
```

```python
# Read the file. Call it df2.
df = pd.read_excel (file, header=0) # header=0 tells pandas that the zeroth row has headers. The counting of rows starts with zero and moves forward as 0,1,2, etc.
df
```

```python
# Sort sequentially based on Adjusted Time.
sorted = df.sort_values(by = ['Time'])
sorted
```

```python
sorted['Distance'] = ((sorted['Position X'] - sorted['Position X'].iloc[0])**2 + (sorted['Position Y'] - sorted['Position Y'].iloc[0])**2 + (sorted['Position Z'] - sorted['Position Z'].iloc[0])**2)**(1/2)
sorted['Adjusted_Time'] = (sorted['Time_seconds'] - sorted['Time_seconds'].iloc[0])
sorted
```

```python
outputfilename = input('Which division round?')
```

```python
sorted.to_excel('Distance data ' + outputfilename + ".xlsx")
```

```python
# Plot... a 2d plot, where the adjusted coordinate values were plotted. 
# x=0, y=0 were the coordinates of the first dividing cells. 
# In the below plot, rows of cells can be imagined with a centre around 0,0.)

from matplotlib import pyplot as plt
import numpy as np

# Define the relevant columns as 'x' and 'y' for convenience.
x = sorted.Distance
y = sorted.Adjusted_Time

# Calculate the Pearson correlation coefficient to determine if x and y have a linear relationship.
r_values = np.corrcoef(x,y) # This gives a 2X2 matrix of r values. We are interested in the values 'between' groups 
# and not the correlation of each group with itself. Run r_values to see how the output looks.

r_string = "{:.2f}".format(r_values[0,1]) # 2f asks the system to limit the number of decimal places in the value in the
# 0th row, 1st column in r_values to 2 digits. Also, in this step, the integer in the location is converted to a string for
# concatenation in the next lines of code.

# Plot a scatter plot.
plt.style.use('fivethirtyeight')
plt.scatter(x, y, marker='o', alpha = 1)
plt.xlabel("Distance (m)", fontsize=14)
plt.ylabel("Timing of cell division (seconds)", fontsize=14)
plt.xlim([-15,400]) # Set lower and upper x limits. -15 so that the dot at '0' doesn't get cut off.
plt.ylim([-15,275]) # Set lower and upper y limits. -15 so that the dot at '0' doesn't get cut off.

# Add the correlation coefficient value as the plot title
plt.title(outputfilename + ' cells' + ': Pearson r = ' + r_string)

# Calculate the slope and y-intercept of the trendline if a trendline is to be plotted.
m,c = np.polyfit(x,y,1) # m = slope, b = intercept

# Add the trendline.
plt.plot(x, m*x + c, 'black', linewidth = 2)

# Add the equation to the plot
m_string = "{:.2f}".format(m) # Convert m to string and limit to 2 decimal places.
c_string = "{:.2f}".format(c) # Convert c to string and limit to 2 decimal places.
plt.text(15, 225, 'y = ' + m_string + 'x + ' + c_string, fontsize = 16)

# Calculate speed by subtracting x when y = 0 s from x when y = 60 s.
x0 = -c/m # Calculate x0 (i.e. the x value when y = 0 s)
x60 = (60-c)/m # Calculate x0 (i.e. the x value when y = 0 s)

speed = x60 - x0
speed_string = "{:.2f}".format(speed)
plt.text(15, 205, speed_string + ' m/min', fontsize = 16)

plt.savefig('Div timing vs distance with trendline ' + outputfilename + '.svg', bbox_inches='tight')
plt.show()
```
