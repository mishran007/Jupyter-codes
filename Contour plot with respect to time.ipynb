```python
import numpy as np
import matplotlib.pyplot as plt

# Import data file.
import pandas as pd
from ipywidgets import *
from tkinter import Tk, filedialog
from IPython.display import clear_output, display
```

```python
# Choose the div output file.

def select_files(b):
    clear_output()                                         # Button is deleted after it is clicked.
    root = Tk()
    root.withdraw()                                        # Hide the main window.
    root.call('wm', 'attributes', '.', '-topmost', True)   # Raise the root to the top of all windows.
    b.files = filedialog.askopenfilename(multiple=False)    # List of selected files will be set button's file attribute.
    print(b.files)                                         # Print the list of files selected.
    
fileselect = Button(description="Select File")
fileselect.on_click(select_files)

display(fileselect)
```

```python
# Call the file in the above location 'file1'. 
file1 = fileselect.files
```

```python
# Read 'file1'. Call it df1.
df1 = pd.read_excel (file1, header=0) # This line reads the 'position' sheet in the excel file. header=1 tells pandas
# that the first row has headers. The counting of rows starts with zero and moves forward as 0,1,2, etc.
df1
```

```python
x1 = df1['Position X']
y1 = df1['Position Y']
z1 = df1['Position Z']
time1 = df1['Time_seconds']

# Add a new column (Adjusted_Time)
min_ticks_1_sec = min(df1['Time_seconds'])
df1['Adjusted_Time'] = df1['Time_seconds'] - min_ticks_1_sec
df1
```

```python
time_ticks_1 = df1.Adjusted_Time.unique()
time_ticks_1
```

```python
xlim_1 = min(time_ticks_1)
ylim_1 = max(time_ticks_1)
```

```python
division_round_1 = input('Which division round?')
imaging_date = input('On what date was this imaging done?')
```

```python
%matplotlib inline
import matplotlib as mpl

plt.figure(figsize=(7.5, 6))
plt.tricontourf(x1, y1, time1, levels = (len(time_ticks_1)-1)) # levels is the number of different colors minus 1 (i.e. # intervals)
plt.scatter(x1,y1, color = 'black', s = z1)
plt.title(division_round_1 + ' cells -' + imaging_date + '  (Dot size = Position Z)')

plt.clim(xlim_1, ylim_1)
plt.xlabel('X-coordinate', fontsize = 15)
plt.ylabel('Y-coordinate', fontsize = 15)
plt.colorbar(ticks = time_ticks_1, label = 'Division Timing (s)')
plt.savefig('Timewise contour plot_' + division_round_1 + '.svg')
plt.show()
```
